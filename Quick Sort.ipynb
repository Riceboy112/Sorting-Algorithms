```python
import sys
import time
import timeit


def timeEfficiency(fileChoose):
    NOPE = " "
    textFile = open(fileChoose, "r")

    timeStart = time.time()
    
    arrays = open(fileChoose).read().split()
    arrays = [int(numbers) for numbers in arrays]
    
    arraySize = len(arrays)
    
    
    print ("\nBefore array is sorted: ", arrays)
    quickSort(arrays, 0, arraySize - 1)
    print("\n\nAfter array is sorted: ", arrays)      
    
    endTime = time.time()
    textFile.close()
    
    timeTaken = endTime - timeStart
    print("\n\nStarts at: ", timeStart)
    print("Ends at: ", endTime)
    print("Time taken to execute function: ", timeTaken)
    return NOPE


def quickSort(array, low, high): 
    if low < high: 
        index = splittingArray(array, low, high)  
        quickSort(array, low, index - 1) 
        quickSort(array, index + 1, high)
        
        
def splittingArray(array, low, high): 
    i = (low - 1)
    pivot = array[high]
    for j in range(low, high): 
        if array[j] <= pivot: 
            i = i + 1 
            array[i], array[j] = array[j], array[i] 
    array[i + 1], array[high] = array[high], array[i + 1] 
    return (i + 1)


print("This is the quick sort sorting algorithm.")
print("We only have one file that has 200 random integers. It will automatically be inputted.")
fileChoose = "rand200.txt"
print(timeEfficiency(fileChoose))
```
