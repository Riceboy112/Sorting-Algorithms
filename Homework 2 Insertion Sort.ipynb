```python
# Function to do insertion sort 
import sys
import timeit
import time


def timeEfficiency(fileChoose):
    NOPE = " "   
    textFile = open(fileChoose, "r")

    timeStart = time.time()
    
    arrays = open(fileChoose).read().split()
    arrays = [int(numbers) for numbers in arrays]
    
    print ("\nBefore array is sorted: ", arrays)
    insertionSort(arrays)
    print("\n\nAfter array is sorted: ", arrays)      
    
    endTime = time.time()
    textFile.close()
    
    timeTaken = endTime - timeStart
    print("\n\nStarts at: ", timeStart)
    print("Ends at: ", endTime)
    print("Time taken to execute function: ", timeTaken)
    return NOPE


def insertionSort(array): 
    for i in range(1, len(array)):
        j = i
        while j > 0 and array[j] < array[j-1]:
            temp = array[j]
            array[j] = array[j - 1]
            array[j - 1] = temp
            j = j - 1
            
#array = []  

print("Choose a file 'rand1000000.txt' or 'rand1000.txt' and I will sort its contents using merge sort.")
fileChoose = input("Enter file name: ") 
print(timeEfficiency(fileChoose))
```
