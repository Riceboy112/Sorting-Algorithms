```python
import sys
import time
import timeit
from math import log10


def timeEfficiency(fileChoose):
    NOPE = " "
    textFile = open(fileChoose, "r")

    timeStart = time.time()
    
    arrays = open(fileChoose).read().split()
    arrays = [int(numbers) for numbers in arrays]
    arraySize = len(arrays)
    
    # Uncomment if you want to see the array before it is sorted
    #if arraySize > 1001:
    #    print ("\nBefore array is sorted: ", arrays)
    
    combinationSort(arrays)
    #print("\n\nAfter array is sorted: ", arrays)      
    
    endTime = time.time()
    textFile.close()
    
    timeTaken = endTime - timeStart
    print("\n\nStarts at: ", timeStart)
    print("Ends at: ", endTime)
    print("Time taken to execute function: ", timeTaken)
    return NOPE


def binarySearch(arrays, elements, start, end):
    if start == end:
        if arrays[start] > elements:
            return start
        else:
            return start + 1
    if start > end:
        return start

    middle = round((start + end)/ 2)

    if arrays[middle] < elements:
        return binarySearch(arrays, elements, middle + 1, end)

    elif arrays[middle] > elements:
        return binarySearch(arrays, elements, start, middle - 1)

    else:
        return middle


def insertionSort(arrays):
    left = len(arrays)
    for index in range(1, left):
        value = arrays[index]
        position = binarySearch(arrays, value, 0, index - 1)
        arrays = arrays[:position] + [value] + arrays[position:index] + arrays[index + 1:]
    return arrays

def merge(left, right):
    if not left:
        return right
    if not right:
        return left
    if left[0] < right[0]:
        return [left[0]] + merge(left[1:], right)
    return [right[0]] + merge(left, right[1:])

def combinationSort(arrays):
    runs, sortedRuns = [], [] #Arrays
    length = len(arrays)
    newRun = [arrays[0]]

    # For every i in the range of 1 to length of array
    for i in range(1, length):
        # If i is at the end of the list
        if i == length - 1:
            newRun.append(arrays[i])
            runs.append(newRun) 
            break

        if arrays[i] < arrays[i - 1]:
            if not newRun:
                runs.append([arrays[i]])
                newRun.append(arrays[i])
            else:
                runs.append(newRun)
                newRun = [arrays[i]]
        else:
            newRun.append(arrays[i])

    # For every item in runs, append it using insertion sort
    for element in runs:
        sortedRuns.append(insertionSort(element))
    
    # For every run in sorted_runs, merge them
    sortedArray = []
    for run in sortedRuns:
        sortedArray = merge(sortedArray, run)

    print(sortedArray)

print("This program is functional, but when I try to use the combination sort for the 1 million data set, it says that the  maximum recursion depth exceeded in comparison.")
print("This is the combination of merge sort and insertion sort sorting algorithm.")
print("Choose a file 'rand1000000.txt', 'rand1000.txt', or 'rand200.txt' and I will sort its contents using merge sort.")
fileChoose = input("Enter file name: ") 
print(timeEfficiency(fileChoose))
```
